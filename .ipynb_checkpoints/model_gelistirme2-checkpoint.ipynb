```python
augmentor = alb.Compose([
    alb.HorizontalFlip(p=0.5),
    alb.VerticalFlip(p=0.5),
    alb.RandomBrightnessContrast(p=0.2),
    alb.RandomGamma(p=0.2),
    alb.RandomRotate90(p=0.5),
    alb.RGBShift(p=0.2),
    alb.RandomCrop(height=450, width=450)],
    bbox_params=alb.BboxParams(format="albumentations", label_fields=["class_labels"]))
```

```python
img = plt.imread("datatest/output/frame_forestsmoke38_0.jpg")
plt.imshow(img)
```

```python
def yolo_to_voc(yolo_coordinates, image_width, image_height):
    x_center, y_center, bbox_width, bbox_height = yolo_coordinates
    x_center *= image_width
    y_center *= image_height
    bbox_width *= image_width
    bbox_height *= image_height
    x_min = int(x_center - (bbox_width / 2))
    y_min = int(y_center - (bbox_height / 2))
    x_max = int(x_center + (bbox_width / 2))
    y_max = int(y_center + (bbox_height / 2))
    voc_coordinates = [x_min, y_min, x_max, y_max]
    voc_coordinates = np.divide(voc_coordinates, [1920,1080,1920,1080])
    return voc_coordinates.tolist()

# Örnek kullanım:

yolo_coordinates = [0.504167, 0.533333, 0.991667, 0.477778]  # Örnek YOLO koordinatları
image_width = 1920  # Görüntü genişliği
image_height = 1080  # Görüntü yüksekliği
voc_coordinates = yolo_to_voc(yolo_coordinates, image_width, image_height)
voc = voc_coordinates

print((voc_coordinates))
```

```python
labels = ["fire", "non-fire"]
image_width = 1920  # Görüntü genişliği
image_height = 1080  # Görüntü yüksekliği
voc_coordinates = yolo_to_voc(yolo_coordinates, image_width, image_height)
for label in labels:
    for image in os.listdir(os.path.join("TRAIN",label)):
        img = cv2.imread(os.path.join("TRAIN", label, image))
        coords = [0,0,0,0]
        
        label_path = os.path.join("datatrain", "labels", f"{image.split('.')[0]}.txt")
        
        try:
            l = []
            annotation = {}
            annotation["image"] = image
            if os.path.exists(label_path):
                with open(label_path, "r") as f:
                    line = f.readline().strip()
                    l.append(line)
                yolo_coordinates = l[0].split(" ")[1:]
                
                yolo_coordinates = [float(val) for val in yolo_coordinates]
                voc_coordinates = yolo_to_voc(yolo_coordinates, image_width, image_height)
                annotation["bbox"] = voc_coordinates
                annotation["class"] = 0
            else:
                annotation["bbox"] = [0,0,0,0]
                annotation["class"] = 1
                
            with open(os.path.join("TRAIN", "train_json", f"{image.split('.')[0]}.json"), "w") as f:
                json.dump(annotation, f)
        
        except Exception as e:
            print(e)
        
        
                
```

```python
train_images = tf.data.Dataset.list_files("TRAIN\\*\\*.jpg", shuffle=False)
train_images = train_images.map(load_image)
train_images = train_images.map(lambda x: tf.image.resize(x, (224,224)))
train_images = train_images.map(lambda x: x/255)
```

```python
def load_labels(label_path):
    with open(label_path.numpy(), 'r', encoding = "utf-8") as f:
        label = json.load(f)

    return [label["class"]], label["bbox"]
```

```python
train_labels = tf.data.Dataset.list_files("train_json\\*.json", shuffle=False)
train_labels = train_labels.map(lambda x: tf.py_function(load_labels, [x], [tf.uint8, tf.float16]))
```

```python
train = tf.data.Dataset.zip((train_images, train_labels))
train = train.shuffle(15)
train = train.batch(8)
train = train.prefetch(4)
```

```python
data_samples = train.as_numpy_iterator()
res = data_samples.next()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx in range(4):
    sample_image = res[0][idx]
    sample_coords = res[1][1][idx]
    print(sample_coords)
    cv2.rectangle(sample_image,
                tuple(np.multiply(sample_coords[:2], [120, 120]).astype(int)),
                tuple(np.multiply(sample_coords[2:], [120, 120]).astype(int)),
                (255, 0, 0), 2)
    
    ax[idx].imshow(sample_image)
```

```python
feature_extractor_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

pretrained_model_without_top_layer = hub.KerasLayer(
    feature_extractor_model, input_shape=(224, 224,3), trainable=False)
```

```python
number_of_classes = 1

model = tf.keras.Sequential([
    pretrained_model_without_top_layer,
    
    tf.keras.layers.Dense(1280, activation="relu"),
    tf.keras.layers.Dense(number_of_classes, activation="sigmoid"),
    
    tf.keras.layers.Dense(1280, activation="relu"),
    tf.keras.layers.Dense(4, activation="softmax")
    model()
])
model.summary()
```
